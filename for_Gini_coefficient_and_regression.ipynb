6. calculate Gini coefficients

```python
def calculate_gini_coefficients(community_data, output_dir): # the function for calculating the Gini coefficients based on UPI for each year from 2014 to 2023.
    """
    Calculate the Gini coefficients based on UPI for each year from 2014 to 2023.
    Parameters:
        community_data (dict): A dictionary containing UPI and total_house_num data for each year.
            The keys are the years (e.g., 2014, 2015), and the values are dictionaries containing the data.
    Returns:
        dict: A dictionary containing the Gini coefficients for each year.
    """
    gini_coefficients = {}
    for year, data in community_data.items():
        # Get UPI values and total house numbers for the current year
        upi_values = data['UPI']
        total_houses = sum(data['total_house_num'])
        
        # Check for NaN values in UPI or 'total_house_num' data
        if np.isnan(upi_values).any() or np.isnan(data['total_house_num']).any():
            print(f"Skipping year {year} due to missing data.")
            continue
        
        # Sort the UPI values in ascending order
        sorted_upi = np.sort(upi_values)
       
        # Calculate the cumulative share of green space proportion
        total_green_space = np.sum(sorted_upi)
        cumulative_green_space = np.cumsum(sorted_upi) / total_green_space
        
        # Calculate the cumulative share of the population proportion (using total_house_num)
        cumulative_houses = np.cumsum(data['total_house_num']) / total_houses
        
        #  Plot the Lorenz curve
        plt.plot(cumulative_houses, cumulative_green_space, label=f'Lorenz Curve {year}')
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Equality Line')
        plt.xlabel('Cumulative Share of House (Families)')
        plt.ylabel('Cumulative Share of Urban Green Space')
        plt.title(f'Lorenz Curve for {year}')
        plt.legend()

        # Save the plot to the disk
        plot_filename = f'{output_dir}/lorenz_curve_{year}.png'
        plt.savefig(plot_filename)
        plt.close()  # Close the plot to free up memory
        
        # Calculate the Gini coefficient
        area_under_lorenz_curve = np.trapz(cumulative_green_space, x=cumulative_houses)
        area_under_equality_line = 0.5
        gini_coefficient = (area_under_equality_line - area_under_lorenz_curve) / area_under_equality_line
        print(f"Gini coefficient for {year}: {gini_coefficient}")
        
        # Store the Gini coefficient in the dictionary
        gini_coefficients[year] = gini_coefficient

    return gini_coefficients

output_directory = 'Gini_result'
gini_coefficients_result = calculate_gini_coefficients(community_data, output_directory)
gini_df = pd.DataFrame(list(gini_coefficients_result.items()), columns=['Year', 'Gini'])
gini_df
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(gini_df['Year'], gini_df['Gini'], marker='o', color='skyblue', label='Gini Coefficient')
plt.xlabel('Year')
plt.ylabel('Gini Coefficient')
plt.title('Gini Coefficients based on UPI between 2014 and 2023')

# Set the y-axis range automatically based on the minimum and maximum Gini values
plt.ylim(min(gini_df['Gini']) - 0.02, max(gini_df['Gini']) + 0.02)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Save the plot
plt.savefig('Gini_result/line_chart_gini_2014_2023.png')

plt.show()
```
